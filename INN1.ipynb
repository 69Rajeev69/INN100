```python
import numpy as np
```

```python
class Neuron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, inputs):
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        output = self.sigmoid(weighted_sum)
        return output
```

```python
input_size = 10
neuron = Neuron(input_size)
inputs = np.random.rand(input_size)
output = neuron.forward(inputs)

print("Input:", inputs)
print("Weights:", neuron.weights)
print("Bias:", neuron.bias)
print("Weighted Sum:",  np.dot(inputs, neuron.weights) + neuron.bias)
print("Output:", output)
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Neuron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, inputs):
        weighted_sum = np.dot(inputs, self.weights) + self.bias
        output = self.sigmoid(weighted_sum)
        return output

input_size = 5
num_samples = 100
inputs = np.random.rand(num_samples, input_size)

neuron = Neuron(input_size)

outputs = np.zeros(num_samples)
for i in range(num_samples):
    outputs[i] = neuron.forward(inputs[i, :])
print(outputs)
x_values = np.linspace(-10, 10, 100)
sigmoid_values = neuron.sigmoid(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, sigmoid_values, label='Sigmoid Activation Function')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()

```
